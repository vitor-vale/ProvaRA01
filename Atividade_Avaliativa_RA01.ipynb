# Questão 1: 
A população de uma cidade está crescendo exponencialmente. No ano de 2020, a população era de 50.000 habitantes, e a taxa de crescimento anual é de 3%. Qual será a população em 2030? Use a fórmula:

$$
P(t) = P_0 \cdot (1 + r)^t
$$

onde $P_0$ é a população inicial, $r$ é a taxa de crescimento e $t$ é o tempo em anos. Não se esqueça de traçar um gráfico mostrando o crescimento populacional entre os anos de 2020 e 2040, ano a ano.


```python
import matplotlib.pyplot as plt
import numpy as np

print("OK")
```

# **Resolução Exercício 1:**

Temos que:


$P(0) = 50.0000$


$r = 0.003$


$P(0) = 50.0000$

Temos também que de 2020 para 2030 a diferença são de 10 anos:


$t = 10$

Substituindo temos: 


$x = 10$


$
P(10) = 50.000 \cdot (1 + 0.003)^x
$

$
P(10) = 51.520,4
$

```python
#Gráfico Exercício 1:

def f(t):
    return 50000 * (1 + 0.003) ** t

t = np.arange(0, 21)
y = f(t)

plt.plot(t, y)
plt.scatter(0, 50000, color='red', s=50)
plt.scatter(10, 51520, color='red', s=50)
plt.scatter(20, 53087, color='red', s=50)
plt.xlabel('Anos')
plt.ylabel('População')
plt.title('Crescimento populacional entre 2020-2040')
plt.show()
```

Usando o Sympy, ou qualquer outro recurso disponível na internet, escreva, com qualidade matemática o passo-a-passo algébrico que leva a determinação dos seguintes limites. Ao final, para cada uma das equações você deverá traçar um gráfico da função. 

1. $\lim_{x \to 2} \frac{x^3 - 8}{x - 2}$

2. $\lim_{x \to \infty} \frac{3x^2 + 2x - 1}{x^2 - 4x + 7}$


# **Reolução:**

**Primeira Equação:**

Temos:

$\lim_{x \to 2} \frac{x^3 - 8}{x - 2}$

Substituindo:

$\lim_{x \to 2} \frac{2^3 - 8}{2 - 2}$

$\lim_{x \to 2} \frac{8 - 8}{2 - 2}$

$\lim_{x \to 2} \frac{0}{0}$

Chegamos em uma indeterminação, logo precisamos manipular algebricamente esse limite. Abrindo o produto notável de cima:

$\lim_{x \to 2} \frac{(x-2)(x^2 +2x + 4)}{x - 2}$

Agora é possível cortar os dois parênteses em comum no numerador e denominador da fração temos:

$\lim_{x \to 2} \frac{(x-2)(x^2 +2x + 4)}{x - 2}$

$\lim_{x \to 2} x^2 +2x + 4$

Substituindo:

$\lim_{x \to 2} 2^2 +2\cdot(2) + 4 = 12$


**Segunda Equação:**

Temos:

$\lim_{x \to \infty} \frac{3x^2 + 2x - 1}{x^2 - 4x + 7}$

Substituindo:

$\lim_{x \to \infty} \frac{3\cdot(\infty)^2 + 2\cdot(\infty) - 1}{\infty^2 - 4\cdot(\infty) + 7}$

$\lim_{x \to \infty} \frac{\infty}{\infty}$

Chegamos em uma indeterminação, logo precisamos manipular algebricamente esse limite. Dividimos todo o Numerador e todo o denominador por $x^2$:

$$\lim_{x \to \infty} \frac{\frac{(3x^2)}{x^2} + \frac{2x}{x^2} - \frac{1}{x^2}}{\frac{x^2}{x^2} - \frac{4x}{x^2} + \frac{7}{x^2}}$$

Temos que $\frac{(3x^2)}{x^2}$ e $\frac{7}{x^2}$ convergem a 0, e que $\frac{x^2}{x^2} = 1$

Ficamos com:

$$\lim_{x \to \infty} \frac{ -\frac{1}{x^2}}{1 - \frac{4x}{x} + \frac{7}{x^2}}$$

O denominador converge para 1 à medida que x tende ao infinito, pois os termos com $\frac{1}{x}$ e $\frac{1}{x^2}$ se tornam insignificantes comparanco com a constante 1, logo temos: 

$$\frac{-1}{1} = -1$$




```python
#Gráficos do Exercício 3:

#Limite 1:
def f(x):
    return x**3-8/x-2

c = 2
L = np.inf

x = np.linspace(-2, 5, 1000)
y = f(x)

plt.plot(x, y, label='x**3-8/x-2')
plt.axvline(x=c, color='gray', linestyle='--', label=f'x = {c}')
plt.axhline(y=L, color='gray', linestyle='--', label=f'y = {L}')
plt.title('Limite 1')
plt.legend()
plt.show()


#Limite 2:
def f(x):
    return 3*x**2+2*x-1/x**2-4*x+7

c = np.inf
L = np.inf

np.linspace(10,100, 1000)
y = f(x)

plt.plot(x, y, label='3*x**2+2*x-1/x**2-4*x+7')
plt.axvline(x=c, color='gray', linestyle='--', label=f'x = {c}')
plt.axhline(y=L, color='gray', linestyle='--', label=f'y = {L}')
plt.title('Limite 2')
plt.legend()
plt.show()

```
